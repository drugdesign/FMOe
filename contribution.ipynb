# Contribution

## Document generation

This documents are generated by Sphinx.

Svl scripts with docstrings are converted into null python scripts with docstrings by svl2py.py.

### Pre-requirement

- install pandoc
    - macOS: brew install pandoc
    - Ubuntu: sudo apt-get install pandoc
- install sphinx and extensions
    - cd $moehome/docs
    - pip install -r requirement.txt

### Build documents

```
cd $moehome/docs
make html
```
